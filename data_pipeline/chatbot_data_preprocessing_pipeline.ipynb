# 6개 카테고리 파일 생성

```python
import urllib.request
import pandas as pd
import csv
from pprint import pprint
```

```python
urllib.request.urlretrieve(
    "https://raw.githubusercontent.com/songys/Chatbot_data/master/ChatbotData.csv",
    filename="ChatBotData.csv",
)
Chatbot_Data = pd.read_csv("ChatBotData.csv")
Chatbot_Data.columns
```

```python
sentiment_data_training = pd.read_excel("/content/drive/MyDrive/KHUDA/심화프로젝트/datasets/AIHub_감성대화/Training_221115_add/원천데이터/감성대화말뭉치(최종데이터)_Training/감성대화말뭉치(최종데이터)_Training.xlsx")
sentiment_data_validation = pd.read_excel("/content/drive/MyDrive/KHUDA/심화프로젝트/datasets/AIHub_감성대화/Validation_221115_add/원천데이터/감성대화말뭉치(최종데이터)_Validation/감성대화말뭉치(최종데이터)_Validation.xlsx")
```

```python
indexed_sentiments_list = list(sentiment_data_training['감정_대분류'].unique())
indexed_sentiments_dict = {}
for i in range(len(indexed_sentiments_list)):
  indexed_sentiments_dict[indexed_sentiments_list[i]] = i
pprint(indexed_sentiments_dict)
print(list(indexed_sentiments_dict.keys()))

chatbot_sentiment_convert_dict = {
    0:6,  # 0(일상다반사) -> 6(일상다반사) 으로 매치해 변환
    1:4,  # 1(이별, 부정) -> 4(슬픔) 으로 매치해 변환
    2:1  # 2(사랑, 긍정) -> 1(기쁨) 으로 매치해 변환
    }

print(list(chatbot_sentiment_convert_dict.keys()))
```

```python
sentiment_data_training_modified = sentiment_data_training[['감정_대분류', '사람문장1', '시스템문장1', '사람문장2', '시스템문장2', '사람문장3', '시스템문장3']]
sentiment_data_validation_modified = sentiment_data_validation[['감정_대분류', '사람문장1', '시스템문장1', '사람문장2', '시스템문장2', '사람문장3', '시스템문장3']]

sentiment_data_training_modified['감정_대분류'] = sentiment_data_training_modified['감정_대분류'].apply(lambda sentiment: indexed_sentiments_dict[sentiment])
sentiment_data_validation_modified['감정_대분류'] = sentiment_data_validation_modified['감정_대분류'].apply(lambda sentiment: indexed_sentiments_dict[sentiment])

Chatbot_Data["label"] = Chatbot_Data["label"].apply(lambda sentiment: chatbot_sentiment_convert_dict[sentiment])
Chatbot_Data.head()
```

```python
chatbot_data_training_1 = sentiment_data_training_modified[['사람문장1', '시스템문장1', '감정_대분류']].dropna(subset=['사람문장1', '시스템문장1', '감정_대분류'])
chatbot_data_validation_1 = sentiment_data_validation_modified[['사람문장1', '시스템문장1', '감정_대분류']].dropna(subset=['사람문장1', '시스템문장1', '감정_대분류'])
chatbot_data_training_1.columns = ['Q', 'A', 'label']
chatbot_data_validation_1.columns = ['Q', 'A', 'label']
chatbot_data_training_2 = sentiment_data_training_modified[['사람문장2', '시스템문장2', '감정_대분류']].dropna(subset=['사람문장2', '시스템문장2', '감정_대분류'])
chatbot_data_validation_2 = sentiment_data_validation_modified[['사람문장2', '시스템문장2', '감정_대분류']].dropna(subset=['사람문장2', '시스템문장2', '감정_대분류'])
chatbot_data_training_2.columns = ['Q', 'A', 'label']
chatbot_data_validation_2.columns = ['Q', 'A', 'label']
chatbot_data_training_3 = sentiment_data_training_modified[['사람문장3', '시스템문장3', '감정_대분류']].dropna(subset=['사람문장3', '시스템문장3', '감정_대분류'])
chatbot_data_validation_3 = sentiment_data_validation_modified[['사람문장3', '시스템문장3', '감정_대분류']].dropna(subset=['사람문장3', '시스템문장3', '감정_대분류'])
chatbot_data_training_3.columns = ['Q', 'A', 'label']
chatbot_data_validation_3.columns = ['Q', 'A', 'label']

Chatbot_Data.dropna(subset=['Q', 'A', 'label'])
```

```python
# Concatenate the DataFrames vertically and ignore the existing index
# merged_chatbot_data = pd.concat([chatbot_data_1, chatbot_data_2, chatbot_data_3], ignore_index = True)
merged_chatbot_data = pd.concat([chatbot_data_training_1, chatbot_data_validation_1, chatbot_data_training_2, chatbot_data_validation_2, chatbot_data_training_3, chatbot_data_validation_3, Chatbot_Data], ignore_index=True)
merged_chatbot_data.info()
```

```python
print(merged_chatbot_data['label'].value_counts())
```

```python
merged_chatbot_data.to_csv('/content/drive/MyDrive/KHUDA/심화프로젝트/datasets/AIHub_감성대화/merged_chatbot_data.csv', index=False)
# merged_chatbot_data.to_excel('/content/drive/MyDrive/KHUDA/심화프로젝트/datasets/AIHub_감성대화/merged_chatbot_data.csv', index=False)
```
